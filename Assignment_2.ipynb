###### Scenario 1: Grocery Shopping List You are creating a program to manage a groceryy shopping list. Users should be able to add items, remove items, and display the current list.

```python
grocery_list = []
no_item = int(input("Enter no item you want to add, remove or display a list: ")) 
for item in range(no_item):
    print("\nif you want to add a item in list press 1: ")
    print("if you want to remove a item in list press 2: ")
    print("if you want to display a list press 3: ")
    print("For exit from list press 4: ")
    choice = input("Enter (1,2,3,4) : ")
    
    
    if choice == "1":
        print("\n\tYou wnat to add an item in a list")
        item = input("Enter item: ")
        grocery_list.append(item)
        print(f"\nAfter adding {item} in a list  ")
        print(grocery_list)
    elif choice == "2":
        print("\n\tYou wnat to remove an item in a list")
        item = input("Enter item: ")
        grocery_list.remove(item)
        print(f"\nAfter removing {item} from a list  ")
        print(grocery_list)
    elif choice == "3":
        print("\n\tYou wnat to display a list")
        print(grocery_list)
    else:
        print("Thank you for making a list")
        break;
```

###### Scenario 2: Student Grades You are managing student grades using a dictionary. You need to calculate the average grade.take atleast 5 Students grades & then calculate the average.

```python
type_of_grade = input("you wnat to see grade of inter or university: ")
no_student = int(input("Enter number of student you want to enter: "))
if type_of_grade == "inter":
    print("\tThis is the result of Inter")
    inter_student = {}
    for a in range(no_student):
        student_name = input("Enter Student name: ")
        student_grade = float(input("Enter Student persentage: "))
        inter_student[student_name] = student_grade
    print(inter_student)
    grade_sum = sum(inter_student.values())
    average_of_grade = grade_sum/len(inter_student)
    print(f"Average of grades is {average_of_grade}%")
else:
    print("\tThis is the result of University")
    uni_Student = {}
    uni_Student_atten = {}
    for a in range(no_student):
        student_name = input("Enter Student name: ")
        student_roll_no = int(input("Enter Student roll number: "))
        student_grade = float(input("Enter Student marks: ")) 
        student_attendence = int(input("Enter attendece: "))
        uni_Student[student_name] = student_grade
        uni_Student_atten[student_roll_no] = student_attendence
        grade_sum = sum(uni_Student.values())
        attendence_sum = sum(uni_Student_atten)
        CGPA = (grade_sum * attendence_sum)/attendence_sum
    print(uni_Student )
    print(uni_Student_atten)
    print(f"Average CGPA is {CGPA}")
```

###### Scenario 4: Password Strength Checker You are creating a program to check the strength of passwords based on certain criteria. • Password should be at least 8 characters long. • Password should contain at least one digit. • Password should contain at least one letter.

```python
password = input("Enter your password: ")

if len(password) < 8:
    print("Password should be at least 8 characters long.")
elif not any(char.isdigit() for char in password):
    print("Password should contain at least one digit.")
elif not any(char.isupper() for char in password):
    print("Password should contain at least one capital letter.")
else:
    print(f"Password is strong.")
```
